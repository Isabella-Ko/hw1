```python
from functools import reduce
import copy
# read into word_list 
f = open('dictionary.words.txt')
words_list = f.read().split('\n')
# print(words_list)

# new_words_list: sorted each letters in a word to match combinations of 16 letters
new_words_list = []
for i in words_list:
    new_i = sorted(i.upper())
    new_i = ''.join(new_i)
    new_words_list.append(new_i)
# print(new_words_list)

# Sorted words in the list - prepare for binary search
sort_nwl = sorted(new_words_list)

# input letters 
input_letters = input('unsolved_letters:')
a = sorted(input_letters.upper())
# print(a)

#Question 1: try to match all 16 letters
suggestions = []
# print(a)
def letter_16(a):
    for word in words_list:
        if a == sorted(word.upper()):
            suggestions.append(word)
#         else: 
#             print('no matches')
    print('suggestions for all 16 letters:', suggestions)
letter_16(a)

# get subsets 
from itertools import permutations, combinations

subsets = []
def substrings(input_letters):
    for i in range(len(input_letters)):
        for p in combinations(input_letters, i+1):
            p_string = ''.join(sorted(p))
            subsets.append(p_string)
#     print(subsets)
substrings(input_letters)

#binary search 
def bin_search(sort_nwl, sub):
    start = 0
    end = len(sort_nwl) - 1  
    
    while start <= end:
        middle = (start+end)//2  
        midpoint = sort_nwl[middle]
        
        if midpoint > sub:
            end = middle- 1
        elif midpoint < sub:
            start = middle + 1
        else:
            return midpoint
        
# match any subsets with the dictionary:
suggestions2 = []
def letter_any(subsets, sort_nwl):
    for sub in subsets: 
        match = bin_search(sort_nwl, sub)
        if match != None:  
            suggestions2.append(match)   
letter_any(subsets,sort_nwl)

#generate all matched words, and best recommendation to get highest scores. 
recom = []
for i in suggestions2: 
    index = new_words_list.index(i)
    real_words = words_list[index]
    recom.append(real_words)

recom1 = copy.deepcopy(recom)

# 'Qu'/'qu' matched set
qu_recom = []
for i in recom:
    if 'Qu' in i or 'qu'in i: 
        qu_recom.append(i)

# ranking 
ref = {'A':1,'B':1,'D':1,'E':1,'G':1,'I':1,'N':1,'O':1,'R':1,'S':1,'S':1,'T':1,'U':1,
      'C':2,'F':2,'H':2,'L':2,'M':2,'P':2,'V':2,'W':2,'Y':2,
      'J':3,'K':3,'Q':3,'X':3,'Z':3}   
        
#deal with 'Qu'/'qu'
if qu_recom != []:
#     print('have Qu')
    scores = []
    for i in qu_recom:
        score = 0
        letter_lst = list(i.upper())
        score = reduce(lambda x, y: x+ref[y], letter_lst, 0)
        scores.append(score)

    index2 = scores.index(max(scores))
    su = qu_recom[index2]
#     print('all matched words(include Qu): ', set(qu_recom))
#     print('best suggestions: ', su)
    
    scores2 = []
    for j in list(recom): 
        if j.count('U') == 1 or j.count('u') == 1:
            recom.remove(j)
        else:
            score = 0
            letter_lst = list(j.upper()) 
            score = reduce(lambda x, y: x+ref[y], letter_lst, 0)
            scores2.append(score)

    index3 = scores2.index(max(scores2))
    su2 = recom[index3]
#     print('all matched words: ', set(recom))
#     print('best suggestions: ', su2)
    
    if max(scores) > max(scores2):
        print('all matched words(include Qu): ', set(qu_recom))
        print('best suggestions: ', su)
    else: 
        print('all matched words: ', set(recom))
        print('best suggestions: ', su2)
    
else:
    scores3 = []
    for v in list(recom1):
        score = 0
        letter_lst = list(v.upper()) 
        score = reduce(lambda x, y: x+ref[y], letter_lst, 0)
        scores3.append(score)
#     print(scores3)
    
    if recom1 == []:
        print('No matching')
    else: 
        index4 = scores3.index(max(scores3))
        su3 = recom1[index4]
        print('all matched words: ', set(recom1))
        print('best suggestions: ', su3)
```
